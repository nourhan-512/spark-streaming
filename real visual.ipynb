```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, explode
import matplotlib.pyplot as plt

# Initialize Spark Session
spark = SparkSession.builder.appName("structured_streaming").getOrCreate()

# Define the input directory where your text files are located
input_path = "./streaming/*.txt"

# Read streaming data
streaming_df = spark.readStream.text(input_path)

# Split and explode words
words_df = streaming_df.select(explode(split(streaming_df.value, " ")).alias("word"))

# Process and visualize data
def visualize_word_count(word_count):
    plt.bar(word_count.keys(), word_count.values())
    plt.show()

# Process the data and call the visualization function
def process_and_visualize(query, epoch_id, df):
    word_count = df.groupBy("word").count().collectAsMap()
    print("Word Count:", word_count)
    visualize_word_count(word_count)

# Apply the processing function to the streaming query
query = words_df.writeStream.foreachBatch(process_and_visualize).start()

try:
    # Wait for the streaming computation to finish or timeout after 60 seconds
    query.awaitTermination(60)
except KeyboardInterrupt:
    # Stop the streaming query gracefully when Ctrl+C is pressed
    query.stop()

# Stop the Spark session
spark.stop()
```
